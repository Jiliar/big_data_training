## Graficaci√≥n con archivos locales

```python
import matplotlib.pyplot as plt
import csv
import os

base_dir = os.getcwd()
ruta = os.path.abspath(os.path.join(base_dir, '..', '..', '..','data', 'chapter2', 'horario.txt'))
grafica_generada = False
counter = 0
with open(ruta) as csv_file:
    csv_reader = csv.reader(csv_file, delimiter=',')
    for row in csv_reader:
        if (row[0]+row[1]+row[2]=='28079004' and row[3]=='12'):
            plt.title("√ìxido de nitr√≥geno: " +row[8]+"/"+row[7]+"/"+row[6])
            hora = 0
            desp = 9
            vs = []
            horas = []
            while hora <= 23:
                if row[desp + 2 * hora + 1]=='V':
                    try:
                        vs.append(float(row[desp + 2 * hora]))
                        horas.append(hora)
                    except ValueError:
                        pass
                hora += 1
            plt.plot(horas, vs)
            plt.show()
            grafica_generada = True
            counter += 1
            if counter >= 5:      # Limitar a 1 gr√°ficas
                break

if not grafica_generada:
    print("No se encontraron datos v√°lidos para graficar.")
```
